## Fidelity Calculations
Here we plot fidelities of the outputs. Since the fidelities can be seen to depend only on the angles $\theta_{1}, \theta_{2}$ (by eliminating the $\theta_{3}$ variable), we find that we obtain the fidelity as a surface. There are two clones and hence two fidelities. Our goal is to consider how the fidelities correspond to probabilities in experiments involving the quantum circuit.

```python
# calculator for coefficients in the clone
def CloneCoeffs(theta0, theta1):
    denom = np.sqrt(1 + (np.tan(theta0)**2) * (np.tan(theta1)**2))
    mu = (np.cos(theta0) * np.cos(theta1) + np.sin(theta0) * np.sin(theta1) * np.tan(theta0) * np.tan(theta1))/denom
    xi = (np.sin(theta0) * np.cos(theta1) - np.sin(theta0) * np.sin(theta1) * np.tan(theta1))/denom
    nu = (np.sin(theta0) * np.sin(theta1) * np.tan(theta1) + np.sin(theta0) * np.cos(theta0))/denom
    return mu, xi, nu
```

```python
# a function to calculate the fidelities of the clones
def fidelities(theta0,theta1):
    mu, xi, nu = CloneCoeffs(theta0, theta1)
    etaA = 2 * mu * nu
    etaB = 2 * mu * xi
    fidA = 0.5 * (1 + etaA)
    fidB = 0.5 * (1 + etaB)
    return fidA, fidB
```

```python
theta0s = np.linspace(0,2*np.pi,100)
theta1s = np.linspace(0,2*np.pi,100)
Theta0, Theta1 = np.meshgrid(theta0s, theta1s)
fidA, fidB = fidelities(Theta0, Theta1)
```

```python
fig = plt.figure(figsize = (9,9))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(Theta0, Theta1, fidA)
plt.show()
```

```python
fig = plt.figure(figsize = (9,9))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(Theta0, Theta1, fidB)
plt.show()
```

## Cloning Experiments in a Circuit

```python
# this is a function that implements the cloning that Eve uses
def EveClones(qc, theta):
    # prepare to clone 
    qc.u(2*theta[0], 0, 0, 1)
    qc.cx(1,2)
    qc.u(2*theta[1], 0, 0, 2)
    qc.cx(2,1)
    qc.u(2*theta[2], 0, 0, 1)
    
    # execute the cloning component with the input qubit (the 0 qubit here)
    qc.cx(0,1)
    qc.cx(0,2)
    qc.cx(1,0)
    qc.cx(2,0)
    
    qc.barrier()
    
    return qc
```

```python
# function that runs 10000 shots and records the probability that Bob (resp. Eve) receives the correct measurement. In this case Alice is sending a qubit in the state |-> state so we expect Bob to measure a 1.
def MeasurementStats(theta0, theta1):
    # returns the probability that Bob returns a 1 (which is the correct measurement result in this case). We use 10000 shots here
    qc = QuantumCircuit(3,2)
    qc.x(0)
    qc.h(0)
    qc.barrier()

    
    theta = np.array([theta0, theta1, np.arctan(np.tan(theta0) * np.tan(theta1))])

    EveClones(qc,theta)

    qc.h(0)
    qc.measure(0,0)
    qc.barrier()

    qc.h(1)
    qc.measure(1,1)

    #qc5.draw()

    shots = 10000 # number of samples used for statistics
    sim = Aer.get_backend('qasm_simulator')
    job = execute(qc, backend = sim, shots = shots)
    
    out00 = job.result().get_counts().get("00")
    out01 = job.result().get_counts().get("01")
    out10 = job.result().get_counts().get("10")
    out11 = job.result().get_counts().get("11")
    
    if out00 == None:
        out00 = 0
    if out01 == None:
        out01 = 0
    if out10 == None:
        out10 = 0
    if out11 == None:
        out11 = 0
    
    BobProb = (out10 + out11)/shots
    EveProb = (out01 + out11)/shots
    return BobProb, EveProb
```

```python
# had trouble getting the measurements to work with the meshgrid structure above
# here we investigate traces of the surface
Bobs = []
Eves = []
for j in range(len(theta1s)):
    BobProb, EveProb = MeasurementStats(np.pi/8, theta1s[j])
    Bobs.append(BobProb)
    Eves.append(EveProb)
```

```python
Bobmeas = np.array(Bobs)
Evemeas = np.array(Eves)
```

```python
plt.figure()
plt.plot(theta1s, Bobmeas)
plt.plot(theta1s, Evemeas)
plt.show()
```
